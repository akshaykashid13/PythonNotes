```python
# Linear Search Algorithm
def linearsearch(A, element):
    position = 0
    flag = False
    while position < len(A) and not flag:
        if A[position] == element:
            flag = True
        else:
            position = position + 1
    return flag
    
A = [10,20,30,40,50,60]
element = int(input("Enter element to find: "))
found = linearsearch(A,element)

if found:
    print("Element is in list")
else:
    print('Nop')
```

```python
# Binary Search
#Mid = (Low + High) / 2
#A[Mid] == Key -> Found
#High = Mid - 1 if it is found in lower half
#Low = Mid + 1 if it is found in higher half
#if low > high -> Not found 
```

```python
# Binary Search without recursion.
def binarysearch(A, element):
    low = 0
    high = len(A) - 1
    while low <= high:
        mid = (low + high) // 2
        if element == A[mid]:
            return True
        elif element < A[mid]:
            high = mid - 1
        else:
            low = mid + 1  
    return False

A = [12,24,36,48,72,96]
element = int(input("Enter element to find: "))
found = binarysearch(A, element)

if found:
    print("Element is in the list")
else:
    print("Nop")
```

```python
# Binary search using recursion
def binaryrecursivesearch(A, Key, low, high):
    if low > high:
        return False
    else:
        mid = (low + high) // 2
        if Key == A[mid]:
            return True
        elif Key < A[mid]:
            return binaryrecursivesearch(A, Key, low, mid-1)
        else:
            return binaryrecursivesearch(A, Key, mid+1, high)
        
A = [15,30,45,60,75]
low = 0
high = len(A) - 1
Key = int(input('Enter element to find '))
found = binaryrecursivesearch(A, Key, low, high)
if found:
    print("Element is in the list")
else:
    print("Nop")
        
```

```python
# Stack
#Operations in stack
#push(object) : insert element
#pop(object) : remove and return element
#top() : returns last inserted element
#len() : returns number of element
#isEmpty() : to check is it empty or not
```

```python
# Stack
class Empty(Exception):
    pass

class ArrayStack(Empty):
    def __init__(self):
        self._data = []
    def length(self):
        return len(self._data)
    def is_empty(self):
        return len(self._data) == 0
    
    def push(self, e):
        self._data.append(e)
        
    def pop(self):
        if self.is_empty():
            raise Empty('Stack is Empty')
        return self._data.pop()
    def top(self):
        if self.is_empty():
            raise Empty('Stack is Empty')
        return self._data[-1]
    
s = ArrayStack()
s.push(10)
s.push(20)
print('Stack: ' , s._data)
print('Length ', s.length())
print('Is Empty ', s.is_empty())
print('Popped ', s.pop())
print('Stack: ' , s._data)
```

```python
# Queue
class Empty(Exception):
    pass

class ArrayQueue(Empty):
    def __init__(self):
        self.data = []
        self.size = 0
        self.front = 0
        
    def length(self):
        return self.size
    def is_empty(self):
        return self.size == 0
    def enqueue(self, e):
        self.data.append(e)
        self.size = self.size + 1
    def dequeue(self):
        if self.is_empty():
            raise Empty ("Queue is Empty")
        value = self.data[self.front]
        self.data[self.front] = None
        self.front = self.front + 1
        self.size = self.size - 1
        return value
    def first(self):
        if self.is_empty():
            raise Empty ("Queue is Empty")
        return self.data[self.front]
    
q = ArrayQueue()
q.enqueue(10)
q.enqueue(20)
print('Queue: ', q.data)
print('Length ', len(q))
print('Dequeue ' q.dequeue())
print('Queue: ', q.data)
```
