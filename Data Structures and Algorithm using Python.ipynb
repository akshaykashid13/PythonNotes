```python
# Linear Search Algorithm
def linearsearch(A, element):
    position = 0
    flag = False
    while position < len(A) and not flag:
        if A[position] == element:
            flag = True
        else:
            position = position + 1
    return flag
    
A = [10,20,30,40,50,60]
element = int(input("Enter element to find: "))
found = linearsearch(A,element)

if found:
    print("Element is in list")
else:
    print('Nop')
```

```python
# Binary Search
#Mid = (Low + High) / 2
#A[Mid] == Key -> Found
#High = Mid - 1 if it is found in lower half
#Low = Mid + 1 if it is found in higher half
#if low > high -> Not found 
```

```python
# Binary Search without recursion.
def binarysearch(A, element):
    low = 0
    high = len(A) - 1
    while low <= high:
        mid = (low + high) // 2
        if element == A[mid]:
            return True
        elif element < A[mid]:
            high = mid - 1
        else:
            low = mid + 1  
    return False

A = [12,24,36,48,72,96]
element = int(input("Enter element to find: "))
found = binarysearch(A, element)

if found:
    print("Element is in the list")
else:
    print("Nop")
```

```python
# Binary search using recursion
def binaryrecursivesearch(A, Key, low, high):
    if low > high:
        return False
    else:
        mid = (low + high) // 2
        if Key == A[mid]:
            return True
        elif Key < A[mid]:
            return binaryrecursivesearch(A, Key, low, mid-1)
        else:
            return binaryrecursivesearch(A, Key, mid+1, high)
        
A = [15,30,45,60,75]
low = 0
high = len(A) - 1
Key = int(input('Enter element to find '))
found = binaryrecursivesearch(A, Key, low, high)
if found:
    print("Element is in the list")
else:
    print("Nop")
```

```python
#Dequeue Oprations
#add_first(object): insert at front of the queue
#add_last(object): insert at end of the queue
#delete_first(): remove from front of the queue
#delete_last(): remove from rear of the queue
#first(): returns element at the front
#last(): returns element at the last
#len(): to check length of the queue
#isEmpty() : whether queue is empty or not 
```

```python
class Empty(Exception):
    pass

class ArrayDeque(Empty):
    def __init__(self):
        self.data = []
        self.front = 0
    def length(self):
        return len(self.data)
    def is_empty(self):
        return len(self.data) == 0
    def first(self):
        if self.is_empty():
            raise Empty("Dequeue is Empty")
        return self.data[self.front]
    def add_first(self, e):
        self.data.insert(self.front,e)
    def add_last(self, e):
        self.data.append(e)
    def delete_first(self):
        if self.is_empty():
            raise Empty("Dequeue is Empty")
        value = self.data.pop(self.front)
        return value
    def delete_last(self):
        if self.is_empty():
            raise Empty("Dequeue is Empty")
        value = self.data.pop()
        return value
    
d = ArrayDeque()
d.add_last(10)
d.add_last(20)
d.add_last(30)
d.add_last(40)
d.add_last(50)
print('Dequeue ', d.data)
print('Deleted first item is ',d.delete_first())
print('Deleted first item is ',d.delete_last())
d.add_first(13)
print('Dequeue ', d.data)
```

```python
# Linked List
class Empty(Exception):
    pass

class LinkedList(Empty):
    class Node:
        slots = 'element', 'next'
        def __init__(self,element,next):
            self.element = element
            self.next = next
        
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
        
    def length(self):
        return self.size
    def is_empty(self):
        return self.size == 0
    def add_first(self,e):
        newest = self.Node(e, None)
        if self.is_empty():
            self.head = newest
            self.tail = newest
        else:
            newest.next = self.head
        self.head = newest
        self.size += 1
        
    def add_last(self,e):
        newest = self.Node(e, None)
        if self.is_empty():
            self.head = newest
            self.tail = newest
        else:
            self.tail.next = newest
        self.tail = newest
        self.size += 1
        
    def add_any(self,e,pos):
        newest = self.Node(e, None)
        thead = self.head
        i = 1
        while i < pos:
            thead = thead.next
            i += 1
        newest.next = thead.next
        thead.next = newest
        self.size += 1
            
    def remove_first(self):
        if self.is_empty():
            raise Empty("Linked list is empty")
        value = self.head.element
        self.head = self.head.next
        self.size -= 1
        if self.is_empty():
            self.tail = None
        return value
    
    def remove_last(self):
        if self.is_empty():
            raise Empty("Linked list is empty")
        thead = self.head
        i = 0
        while i < len(self) - 2:
            thead = thead.next
            i = i + 1
        self.tail = thead
        thead = thead.next
        value = thead.element
        self.tail.next = None
        self.size -= 1
        return value
    
    def remove_any(self, pos):
        if self.is_empty():
            raise Empty("Linked list is empty")
        thead = self.head
        i = 1
        while i < pos-1:
            thead = thead.next
            i = i + 1
        value = thead.next.element
        thead.next = thead.next.next
        self.size -= 1
        return value
    
    def display(self):
        thead = self.head
        while thead:
            print(thead.element, end='-->')
            thead = thead.next
        print()
        
L = LinkedList()
L.add_last(10)
L.add_last(20)
L.add_last(30)
L.add_last(40)
L.display()
L.remove_first()
L.display()
L.add_first(100)
L.display()
L.add_any(50,3)
L.display()
L.remove_any(3)
L.display()
```

```python
# Selection Sort
def selectionsort(A):
    for i in range(len(A)-1, 0, -1):
        max_position = 0
        for j in range(1, i + 1):
            if A[j] > A[max_position]:
                max_position = j
        temp = A[i]
        A[i] = A[max_position]
        A[max_position] = temp
        
A = [84,25,66,98,62]
print('Unsorted', A)
selectionsort(A)
print('Sorted array', A)
```

```python
# Insertion Sort
def insertionsort(A):
    for i in range(1, len(A)):
        value = A[i]
        position = i
        
        while position > 0 and A[position - 1] > value:
            A[position] = A[position - 1]
            position = position - 1
        A[position] = value
        
A = [84,25,66,98,62]
print('Unsorted', A)
insertionsort(A)
print('Sorted array', A)
```

```python
# Bubble sort
def bubblesort(A):
    for i in range(len(A)-1, 0, -1):
        for j in range(i):
            if A[j] > A[j+1]:
                A[j], A[j+1] = A[j+1], A[j]
                
A = [84,25,66,98,62]
print('Unsorted', A)
bubblesort(A)
print('Sorted array', A)     
```

```python
# Merge Sort
def mergesort(A):
    if len(A) > 1:
        mid = len(A) // 2
        left = A[:mid]
        right = A[mid:]
        
        mergesort(left)
        mergesort(right)
        
        i = 0
        j = 0
        k = 0
        
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                A[k] = left[i]
                i = i + 1
            else:
                A[k] = right[j]
                j = j + 1
            k = k + 1   
        while i < len(left):
            A[k] = left[i]
            i = i + 1
            k = k + 1
            
        while j < len(right):
            A[k] = right[j]
            j = j + 1
            k = k + 1
            
A = [84,25,66,98,62]
print('Unsorted', A)
bubblesort(A)
print('Sorted array', A) 
```

```python
# Quick Sort
def quicksort(A, low, high):
    if low < high:
        p = partition(A, low, high)
        quicksort(A, low, p-1)
        quicksort(A, high, p+1)
        
def partition(A, low, high):
    i = low - 1
    pivot = A[high]
    for j in range(low, high):
        if A[j] <= pivot:
            i = i + 1
            A[i], A[j] = A[j], A[i]
            
    A[i+1],A[high] = A[high], A[i+1]
    
    return i + 1

A = [84,25,66,98,62]
print('Unsorted', A)
bubblesort(A)
print('Sorted array', A) 
```
